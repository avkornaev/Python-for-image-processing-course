Import libraries.

```python
import cv2
import numpy as np
from scipy.ndimage.interpolation import zoom
from matplotlib import pyplot as plt
import czifile
import imagecodecs
#import tifffile
```

Download "czi" image and create slices of the image

```python
img = czifile.imread('/content/drive/MyDrive/Cells_count/B16_1122_2_2.czi')
print('Image shape: ',img.shape)# ?, ?, ?, number of chanels, x size, y size, ?
img1 = img[0,0,0,:,0,:,:,0]
img2 = img1[0,:,:] # 0 filter
img3 = img1[1,:,:] # 1 filter

plt.figure(figsize=(15,15))
plt.imshow(img3,cmap='Blues')
```

Visualize the image slices

```python
plt.figure(figsize=(15,15))
plt.subplot(1,2,1)
plt.title('H3342-T2 channel highlights all the cells')
plt.imshow(img3)

plt.subplot(1,2,2)
plt.title('fluo4-T1 channel highlights the apoptosis cells')
plt.imshow(img2)


```
