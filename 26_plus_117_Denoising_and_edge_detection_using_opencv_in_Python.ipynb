```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
from matplotlib import pyplot as plt
```

**Bluring is the way to filter images**

```python
img = cv2.imread('/content/drive/MyDrive/BloodCount/img1_00000_000000000000148.png')
kernel = np.ones((5,5), np.float32)
kernel = kernel/np.size(kernel) # normalisation
filt_2D = cv2.filter2D(img,-1,kernel)
gaussian_blur = cv2.GaussianBlur(img, (5,5), 0)
median_blur = cv2.medianBlur(img,3)
bilateral_blur = cv2.bilateralFilter(img,9,75,75)
```

```python
cv2_imshow(img)
cv2_imshow(filt_2D)
cv2_imshow(gaussian_blur)
cv2_imshow(median_blur)
cv2_imshow(bilateral_blur)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
edges = cv2.Canny(bilateral_blur[:,:,1],0.1,2)
cv2_imshow(bilateral_blur[:,:,1])
cv2_imshow(edges)
```
